```python
import turtle
t = turtle.Turtle()
t.pensize(2)
turtle.bgcolor("lightgreen")


## vẽ mặt trời
t.speed(0)
t.penup()
t.goto(150,200)
t.pendown()
r = 30

t.fillcolor("yellow")
t.begin_fill()
for i in range(4):
    t.fd(30)
    t.bk(30)
    t.lt(90)
    t.circle(r,45)
    t.rt(90)
    t.fd(15)
    t.bk(15)
    t.lt(90)
    t.circle(r,45)
    t.rt(90)
t.end_fill()

## vẽ nhà
t.penup()
t.goto(-200,0)
t.pendown()

r = 50 ## cạnh mái nhà
w = 88 ## chiều ngang nhà
h = 120 ## chiều cao nhà

t.fillcolor("pink")
t.begin_fill()
t.fd(w)
t.lt(150)
t.fd(r)
t.lt(60)
t.fd(r)
t.lt(60)
t.end_fill()

t.fillcolor("blue")
t.begin_fill()
for i in range(2):
    t.fd(h)
    t.lt(90)
    t.fd(w)
    t.lt(90)
t.end_fill()   

t.lt(90)
t.fd(w)
t.rt(90)
t.fillcolor("yellow")
t.begin_fill()
for i in range(2):
    t.fd(h)
    t.lt(110)
    t.fd(w)
    t.lt(70)
t.end_fill()

t.fillcolor("orange")
t.begin_fill()
t.lt(110)
t.fd(w)
t.lt(130)
t.fd(r)
t.lt(50)
t.fd(w)
t.end_fill()

t.penup()
t.goto(-200,0)
t.setheading(-90)
t.fd(h)
t.lt(90)
t.fd(w/3)

t.pendown()
t.fillcolor("green")
t.begin_fill()
for i in range(2):
    t.fd(w/3)
    t.lt(90)
    t.fd(h/3)
    t.lt(90)
t.end_fill()

t.penup()
t.goto(-200,0)
t.setheading(0)
t.fd(h/2)
t.rt(35)
t.fd(h/2)
t.setheading(-90)
t.pendown()

t.fillcolor("brown")
t.begin_fill()
for i in range(2):
    t.fd(h/5)
    t.lt(110)
    t.fd(h/5)
    t.lt(70)
t.end_fill()

## vẽ cây thông

t.setheading(0)
t.penup()
t.goto(200,-200)
t.pendown()

t.fillcolor("brown")
t.begin_fill()
for i in range(2):
    t.fd(16)
    t.lt(90)
    t.fd(35)
    t.lt(90)
t.end_fill()

t.penup()
t.goto(208,-200)
t.lt(90)
t.fd(35)
t.rt(90)
t.pendown()

x = 100 ## tán cây thông

for i in range(3):
    t.bk(x/2)
    t.fillcolor("green")
    t.begin_fill()
    for i in range(3):
        t.fd(x)
        t.lt(120)
    t.end_fill()
    t.penup()
    t.lt(60)
    t.fd(x-5)
    t.pendown()
    t.rt(60)
    


```
